2) La Agencia Nacional de Estadísticas de Buenos Aires recolecta
información de nacimientos cuando los padres registran a sus hijos en
el registro civil a partir de una encuesta. Esa información se encuentra
disponible para su análisis en un csv con el siguiente formato
(dia_nacimiento, mes_nacimiento, anio_nacimiento, peso_al_nacer,
longitud_al_nacer, id_hospital, tipo_parto), donde el tipo de parto 1 es
natural y 2 es cesárea.
Por otro lado la agencia cuenta con información histórica de los
hospitales en otro csv con siguiente formato (id_hospital, dirección,
promedio_nacimientos_mensual).
Se pide usar Pandas para:
a) Calcular la cantidad de nacimientos para cada uno de los hospitales
para el mes de Octubre de 2017 e indicar aquellos hospitales que
superan el promedio de nacimientos mensuales.
b) Comparando el mes de Octubre de 2017 indicar programáticamente
si se incremento el % de cesáreas con respecto a ese mes del año 2016.
(****) (15 pts)

```python
nacOct=df_a[(df_a['mes_nacimiento']=='1') & (df_a['anio_nacimiento']=='2017')]
nacOct
```

. b) Comparando el mes de Octubre de 2017 indicar programáticamente si se incremento el % de cesáreas con respecto a ese mes del año 2016. (**) (15 pts)

```python
nacimientos_data2 = {
        'mes_nacimiento': ['1', '2', '3', '1', '2','1','2','3','4','1','2','2','1','1','2','2','2','3','1','1'],
        'anio_nacimiento':['2017','1994','1993','2017','1995','2016','1992','1994','1990','2017','11','12','2016','2017','2016','2016','2016','2016','2017','2017'],
        'id_hospital':[1,2,1,2,2,2,3,1,3,1,2,3,1,2,1,1,2,3,1,1],
        'tipo_parto':[1,1,2,2,2,2,1,1,2,2,1,1,2,1,2,1,2,2,1,2]
}

nacimientos = pd.DataFrame(nacimientos_data2, columns = ['mes_nacimiento','anio_nacimiento','id_hospital','tipo_parto'])
```

```python
nacimientos
```

```python
nacimientos16y17= nacimientos[(nacimientos['anio_nacimiento']=='2016') | (nacimientos['anio_nacimiento']=='2017')]
```

```python
nacimientos16y17
```

```python
nacimientos16y17= nacimientos16y17[['anio_nacimiento','tipo_parto']]
```

```python
nacimientos16y17
```

Opcion1)

```python
group = nacimientos16y17.groupby(['anio_nacimiento']).apply(lambda x: x[x['tipo_parto']==2].count()/x['tipo_parto'].count())['tipo_parto']
```

```python
group
```

```python
aumento = "si" if (group[1] > group[0]+0.1*group[0]) else "no"
```

```python
aumento
```

Opcion2)

```python
nacPorTipos = nacimientos16y17.groupby('anio_nacimiento')['tipo_parto'].value_counts().reset_index(name='cantidad')
```

```python
nacPorTipos
```

```python
nacPorTipos = nacPorTipos.groupby(['anio_nacimiento']).apply(lambda x: x[x['tipo_parto']==2]['cantidad']/x['cantidad'].sum()).reset_index(name='promedio')
```

```python
nacPorTipos
```

```python
aumento = "si" if (nacPorTipos['promedio'][1] > (nacPorTipos['promedio'][0]+0.1*nacPorTipos['promedio'][0])) else "no"
```

```python
aumento
```
